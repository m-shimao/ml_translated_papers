## Abstract

画像のノイズ除去、超解像度、修復などを含む画像修復は、コンピュータビジョンおよび画像処理における十分に研究された問題であり、低レベルの画像モデリングアルゴリズムのためのテストベッドでもある。
本研究では、対称畳み込みデコンボリューション層を持つ符号化復号化フレームワークである、画像修復のための非常に深い完全畳み込み自己エンコーダネットワークを提案する。
言い換えれば、ネットワークは、畳み込み演算子と畳み込み演算子の複数のレイヤーから構成され、破損した画像から元のものへのエンドツーエンドのマッピングを学習します。

畳み込みレイヤは、破損を排除しながら画像コンテンツの抽象化を捕捉する。
デコンボリューション層には、フィーチャマップをアップサンプリングして画像の詳細を復元する機能があります。
より深いネットワークが訓練するのがより困難になる傾向にある問題に対処するために、畳み込みレイヤとデコンボリューションレイヤをスキップレイヤ接続と対称にリンクすることを提案します。
畳み込みレイヤからミラーリングされた対応するデコンボリューションレイヤへのスキップ接続には、2つの主な利点があります。
第1に、信号を直接最下層に逆伝搬させることができ、勾配が消えるという問題に対処し、深いネットワークのトレーニングを容易にし、復旧性能を向上させることができる。
第2に、これらのスキップ接続は、畳み込みレイヤーからデコンボリューションレイヤーに画像の詳細を渡すので、きれいな画像を回復するのに有益です。重要なのは、大容量で、単一のモデルを使用してさまざまなレベルの破損に対処することが可能であることを示しています。

同じフレームワークを使用して、画像ノイズ除去、超解像JPEG圧縮アーチファクト、ノンブラインド画像ブレ除去および画像修復のタスクに関するモデルを訓練する。
Benchmarkのデータセットに関する実験結果は、当社のネットワークが4つのタスクのすべてで最高のパフォーマンスを達成し、新しい最先端技術を確立できることを示しています。

## 2 RELATED WORK

過去数十年間に、様々な画像復元方法を開発するために徹底的な研究が行われてきた。

調査[29]の詳細レビューを参照してください。
BM3Dアルゴリズム[30]や辞書学習ベースの手法[31]、[16]、[32]のような従来の方法は、
画像ノイズ除去や超解像などの画像修復課題において非常に有望な性能を示している。

画像修復の不都合な性質のために、正則化技術として策定された画像事前知識が広く使用されている。
トータルバリエーション[33]、[34]、[35]などのクラシック正則化モデルは、ノイズアーチファクトを除去するのに効果的ですが、
画像をオーバースムーズにする傾向もあります。

代替として、疎な表現[36]、[3]、[37]、[38]、[39]に基づく事前モデリングも普及しています。 
数学的には、疎な表現モデルは、データ点xを過度に複雑な辞書によって線形的に再構成することができ、係数の大部分はゼロまたはゼロに近いと仮定している。

画像復元のためのアクティブな（そしておそらくより有望な）カテゴリは、ニューラルネットワークベースの方法である。
ニューラルネットワーク方法と他の方法との最も重要な相違点は、予め定義された画像プリオリに頼るのではなく、
訓練データ（例えば、きれいで壊れた画像の対）から直接的に画像復元のためのパラメータを学習することである。

Stacked denoising auto-encoder [24]は、画像ノイズ除去に使用できる最もよく知られているディープニューラルネットワークモデルの1つです。
入力に対する再構成誤差を最小にする教師なし事前トレーニングは、一度に1つの層に対して行われる。
すべてのレイヤーが事前トレーニングされると、ネットワークは微調整段階を経ます。

Xieら[18] イメージノイズ除去やインペインティングなどの低レベルのビジョンタスク用のノイズ除去自動エンコーダで事前トレーニングされた疎なコーディングとディープネットワークを組み合わせたものです。
主な考え方は、改善されたパフォーマンスのために、正規化の希薄性を誘導する用語が提案されていることです。
ディープ・ネットワーク・カスケード（Deep Network Cascade：DNC）[40]は、画像超解像のための複数の積み重ねられた共同ローカル・オートエンコーダーのカスケードです。
高周波テクスチャ強調画像パッチがネットワークに供給され、雑音を抑制し、重なり合うパッチの互換性を協調させる。

多層パーセプトロンのようなネットワークを用いた他のニューラルネットワークに基づく画像復元法。
マルチレベルシグモイド関数をもつマルチレイヤパーセプトロン[41]のような初期の研究が提案され、画像修復作業に有効であることが証明されている。

Burgerら[42]は、 プレーンなマルチレイヤパーセプトロンで大規模なデータセットで学習されたパッチベースのアルゴリズムを発表し、BM3Dなどの最先端の従来の画像ノイズ除去手法と競合することができました。

彼らはまた、大規模なネットワークでは、大規模な訓練データ、ニューラルネットワークが最先端の画像ノイズ除去性能を達成できると結論付けました。

自動符号化および多層パーセプトロンと比較して、畳み込みニューラルネットワークは、画像復元の分野においてさらに重要な成功を収めているようである。
Jain and Seung [26]は雑音除去のために完全な畳み込みCNNを提案した[26]。
クリーンイメージとその壊れたバージョンの間の損失を最小限に抑えることにより、ネットワークに騒音を加えることによってネットワークを訓練します。
彼らは、CNNがブラインドとノンブラインドの画像ノイズ除去に優れており、ウェーブレットやマルコフランダムフィールド（MRF）法に匹敵する、
あるいはさらに優れた性能を提供することを発見しました。

最近、Dongら[28]は、 画像超解像度の低/高解像度画像間のエンド・ツー・エンド・マッピングを直接学習することを提案した。

彼らは、畳み込みニューラルネットワークが疎コードベースの方法と根本的に関連していること、
すなわち、ネットワーク内の3つの層をパッチ表現抽出器、非線形マッピングおよび画像再構成器とみなすことができることを観察した。
彼らはまた、JPEGデブロッキング[21]のような他の画像復元タスクのための変形ネットワークを提案した。

Wangら [17]は、従来のスパース符号化によって表現された領域専門知識は依然として価値があり、
画像超解像においてさらに改善された結果を得るために組み合わせることができると主張した。 

スケールファクタの異なるレベルでトレーニングする代わりに、カスケード構造を使用して、
所望のサイズに達するまで低解像度画像を固定スケールで繰り返し拡大することを提案した。

一般に、DNNベースの方法は、実際の画像復元アプリケーションにおいてより効果的であったデータから復元パラメータを直接学習する。

## VERY DEEP CONVOLUTIONAL AUTO-ENCODER FOR IMAGE RESTORATION

提案されたフレームワークは、図1に示すように、主に畳み込みレイヤーと対称デコンボリューションレイヤーのチェーンを含んでいます。

スキップ接続は、畳み込みレイヤーからデコンボリューションレイヤーに対称的に接続されています。
非常に深い残留エンコーダ/デコーダネットワークを「RED-Net」と呼んでいます。

### 3.1 Architecture

このフレームワークは完全に畳み込み（およびデコンボリューションです。デコンボリューションは基本的にサンプリングしない畳み込みです）。
整流層は、各畳み込みとデコンボリューションの後に追加されます。 
低レベルのイメージ復元の問題では、通常、プーリングはこれらのタスクに不可欠な有用なイメージの詳細を破棄するため、ネットワークでプーリングもプールもしません。

畳み込み層とデコンボリューション層が対称であるため、ネットワークは本質的にピクセル単位の予測であり、
したがって入力画像のサイズは任意であることに言及することは有益である。 
ネットワークの入力および出力は、同じサイズw×h×cの画像であり、w、hおよびcは、幅、高さおよびチャンネルの数である。

私たちの主な考え方は、畳み込みレイヤーが画像内のオブジェクトの主成分を保存しながら、
腐敗を排除する特徴抽出器として機能することです。

畳み込み層を介して転送した後、破損した入力画像は「クリーン」な画像に変換される。
この過程で、画像内容の微妙な細部が失われる可能性があります。
その後、デコンボリューション層を組み合わせて画像内容の詳細を復元する。

デコンボリューション層の出力は、復元されたクリーンバージョンの入力画像である。

さらに、畳み込みレイヤから対応するミラー化されたデコンボリューションレイヤにスキップ接続を追加します。

渡された畳み込み特徴マップは、要素ごとに逆畳み込み特徴マップに合計され、整流後に次のレイヤに渡される。

上記のアーキテクチャに基づいて、画像ノイズ除去、画像超解像、
JPEGデブロッキングおよび画像修復のために、それぞれ20層および30層である本発明者らの実験で2つのネットワークを使用した。

### 3.2 Deconvolution decoder

近年、セマンティックセグメンテーションのために畳み込みとデコンボリューションの層を組み合わせたアーキテクチャ[43]、[44]が提案されている。 

デコンボルーションレイヤは、フィルタウィンドウ内の複数の入力アクティベーションを融合して単一のアクティベーションを出力する畳み込みレイヤとは異なり、単一の入力アクティベーションを複数の出力に関連付けます。
デコンボリューションは、通常、学習可能なアップサンプリング層として使用されます。 
私たちのネットワークでは、畳み込みレイヤーは入力画像の内容を小さなサイズの抽象化に連続してダウンサンプリングします。 
Deconvolutionalレイヤーは抽象化を元の解像度にアップサンプリングします。

スキップ接続を使用する以外にも、我々のモデルと[43]、[44]の主な違いは、
私たちのネットワークが完全に畳み込みでデコンボリューション、すなわちプーリングとアンプーリングなしであることです。

その理由は、低レベルのイメージ修復の目的は、イメージ抽象を学習するのではなく、イメージの詳細を保存しながら、
低レベルの破損を排除することにあります。

セグメンテーションや認識などの高度なアプリケーションとは異なり、プーリングは通常、豊富なイメージの詳細を削除し、
復元のパフォーマンスを低下させる可能性があります。

デコンボリューションを単にコンボリューションに置き換えることができ、
最近提案された非常に深い完全畳み込みニューラルネットワーク[27] [28]に非常に類似したアーキテクチャをもたらす。

しかし、完全畳み込みモデルと我々のモデルとの間には本質的な相違が存在する。
例として画像のノイズ除去を行います。
5層と10層の完全畳み込みネットワークをネットワークと比較します（畳み込みと逆畳み込みを組み合わせたものですが、スキップ接続はありません）。
完全畳み込みネットワークでは、入力と出力を同じサイズにするためにパディングまたはアップサンプリングを使用します。

私たちのネットワークでは、最初の5つのレイヤーは畳み込みで、後の5つのレイヤーはデコンボリューションです。
トレーニング用の他のパラメータはすべて同じです（SGDと学習率10-6、騒音レベルσ= 70）。
妥当性検証セットのピーク信号対ノイズ比（PSNR）が報告されています。

デコンボリューションは、図2に示すように完全畳み込みのカウンターパートより優れています。

さらに、図3では、10層完全畳み込みネットワークと我々の層からの層2,5,8,10の出力であるいくつかの結果を視覚化する。
完全コンボリューションの場合、ノイズは段階的に除去される。

すなわち、各層の後にノイズレベルが低減される。この処理中に、画像コンテンツの詳細が失われる可能性があります。

それにもかかわらず、私たちのネットワークでは、コンボリューションは主要画像コンテンツを保存します。 
その後、デコンボリューションを使用して詳細を補正します。

### 3.3 Skip connections

直感的な質問は、画像の抽象化のみから画像の詳細を復元できるデコンボリューションを持つネットワークですか？ 
畳み込みレイヤーを数レイヤしか持たない浅いネットワークでは、デコンボリューションが詳細を回復することができます。 
しかし、デコンボリューションレイヤーを使用していても、ネットワークが深くなったり、最大プールなどの操作を行ったりすると、
うまく機能しない可能性があります。
畳み込みやプールであまりにも多くの詳細が失われている可能性があります。

もう一つの質問は、私たちのネットワークが深くなるとパフォーマンスが向上するということです。
我々は、画像修復タスクにおけるより深いネットワークは、容易に性能低下を被る傾向があることを観察する。 
理由は2つ折りである可能性があります。 
まず第一に、畳み込みの層が増えると、かなりの量の画像の詳細が失われたり壊れたりする可能性があります。 
画像の抽象化のみが与えられているため、その詳細を復元することは未決定の問題です。
第2に、最適化の点で、ディープネットワークは勾配が消えて訓練するのがずっと困難になることがよくあります。
これは、ニューラルネットワークの文献でよく取り上げられている問題です。

図1に示すように、2つの対応する畳み込みレイヤとデコンボリューションレイヤの間にスキップ接続を追加します。
ビルディングブロックを図4に示します。 そのような接続を使用する2つの理由があります。 
第1に、上記のようにネットワークが深くなると、画像の詳細が失われてデコンボリューションが弱くなり、復元が弱くなります。
ただし、スキップ接続によって渡されたフィーチャマップには、多くのイメージの詳細が含まれているため、
デコンボリューションによって改善されたクリーンバージョンのイメージが復元されます。

第2に、スキップ接続は、グラジエントをボトムレイヤーにバックプロパゲーションすることでもメリットが得られます。

[45] [1]のように、より深いネットワークのトレーニングがはるかに容易になります。
スキップ層の接続は[45]と[1]で提案されたスキップ層の接続とは大きく異なります。 
我々の場合、畳み込み特徴マップの情報を対応するデコンボリューション層に渡したいと思う。 
非常に深いハイウェーネットワーク[45]は、本質的に、忘却ゲートを備えたフィードフォワードロング短期記憶（LSTM）であり、
深い残存ネットワークのCNN層は、ゲートのないフィードフォワードLSTMである。 
私たちのネットワークは一般的に標準フィードフォワードLSTMの形式ではないことに注意してください。

入力Xから出力Yへのマッピングを直接学習する代わりに、F（X）= Y-Xと表される問題の残差[1]にネットワークを適合させたい。
そのような学習戦略が適用される エンコード・デコード・ネットワークの内部ブロックに送信し、トレーニングをより効果的にする。 
スキップ接続は、2つの畳み込みレイヤごとに、ミラー化されたデコンボリューションレイヤに渡されます。 
他の構成も可能であり、我々の実験はこの構成が既に非常にうまく動作していることを示している。 
このようなショートカットを使用すると、ネットワークの訓練が容易になり、
ネットワークの深度を増やすことで復元パフォーマンスを向上させることができます。
